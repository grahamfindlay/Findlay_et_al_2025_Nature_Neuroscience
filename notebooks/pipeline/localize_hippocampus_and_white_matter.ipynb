### Using the selection from above, now run a 1D kCSD on whole probe

```python
import matplotlib.pyplot as plt

from ecephys import sglxr
from ecephys import wne
from ecephys import xrsig
from ecephys.signal.utils import mean_subtract
from findlay2025a import core
import wisc_ecephys_tools as wet
```

```python
# For brand-new subjects, we may need to refresh the cache.
refresh_cache = True
subjLib = wne.sglx.SubjectLibrary(wet.get_subjects_directory())
subj = subjLib.get_subject(SUBJECT)
if refresh_cache:
    subj.refresh_cache()
```

```python
# This no longer runs, because the alias is deprecated and no longer exists!
lfpTable = subj.get_lfp_bin_table(EXPERIMENT, ALIAS, probe=PROBE)
lfp_path = lfpTable.path.iloc[fidx]
print(lfp_path)
```

```python
# Load the LFP data
DURATION = 2.0
lfp = sglxr.load_trigger(
    lfp_path, start_time=(START_TIME), end_time=(START_TIME + DURATION)
)
```

```python
nb = core.get_wne_scratch_project()  # This would now be sh.get_project("seahorse")
s3 = core.get_wne_backup_project()  # This would now be sh.get_project("s3")
opts = s3.load_experiment_subject_json(
    EXPERIMENT, SUBJECT, wne.constants.EXP_PARAMS_FNAME
)  # This is now where recovery sleep onset times are specified.

lfp = xrsig.NPX1LFPs(lfp)
lfp.dephase()
lfp.interpolate(opts["probes"][PROBE]["badChannels"])
# lfp = lfp.decimate(q=2)
```

```python
lfp.values = mean_subtract(lfp.values)
```

```python
um_per_mm = 1000
csd = xrsig.NPX1LFPs(lfp).kCSD(doLCurve=True)
```

```python
print(f"Hippocampus: {hpc_lo}um, {hpc_hi}um")
maskHC = (hpc_lo <= csd.y.values) & (hpc_hi >= csd.y.values)
chansHC = csd["channel"].values[maskHC]
print(f"Channels:\n {chansHC}")
print("")

print(f"White matter: {wm_lo}um, {wm_hi}um")
maskWM = (wm_lo <= csd.y.values) & (wm_hi >= csd.y.values)
chansWM = csd["channel"].values[maskWM]
print(f"Channels:\n {chansWM}")
```

```python
# Use this plot to assess the localization. Then go back to the top and update values as needed.
fig, ax = plt.subplots(figsize=(20, 10))
ax.pcolormesh(csd.time.values, csd.kcsd.estm_x, csd.values.T, shading="gouraud")
ax.set_xlabel("Time [sec]")
ax.set_ylabel("Depth (mm)")
ax.set_title(
    f"{SUBJECT}, {EXPERIMENT}.    Source: {lfp_path.stem}, {START_TIME}-{START_TIME + DURATION}s. \nHC: {hpc_lo}-{hpc_hi}um, LF{chansHC[0]}-LF{chansHC[-1]} \nWM: {wm_lo}-{wm_hi}um, LF{chansWM[0]}-LF{chansWM[-1]}"
)
ax.axhline(hpc_hi / um_per_mm, alpha=0.5, color="dimgrey", linestyle="--")
ax.axhline(hpc_lo / um_per_mm, alpha=0.5, color="dimgrey", linestyle="--")
ax.axhline(wm_hi / um_per_mm, alpha=0.5, color="brown", linestyle="--")
ax.axhline(wm_lo / um_per_mm, alpha=0.5, color="brown", linestyle="--")
```
