```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns

from findlay2025a import core
from findlay2025a.constants import Files
```

```python
nb = core.get_project("seahorse")
cxspr = pd.read_parquet(nb.get_project_file(Files.CORTICAL_SPINDLE_CONDITION_RATES))
cxspm = pd.read_parquet(nb.get_project_file(Files.CORTICAL_SPINDLE_CONDITION_MEANS))
hcspr = pd.read_parquet(nb.get_project_file(Files.HIPPOCAMPAL_SPINDLE_CONDITION_RATES))
hcspm = pd.read_parquet(nb.get_project_file(Files.HIPPOCAMPAL_SPINDLE_CONDITION_MEANS))
measures = pd.concat([cxspm, cxspr, hcspm, hcspr], axis="columns")
df = measures.reset_index()
```

```python
import warnings

warnings.filterwarnings("ignore", category=PendingDeprecationWarning)
```

```python
p = sns.color_palette("Paired")
for i in np.arange(len(p))[1::2]:  # Desaturate some colors a bit
    p[i] = tuple((np.array(p[i]) + np.array(p[i - 1])) / 2)

p = {
    k: v
    for k, v in zip(
        [
            "early_bsl_nrem",
            "early_rec_nrem_match",
            "early_ext_wake",
            "late_ext_wake",
            "early_rec_nrem",
            "late_rec_nrem",
        ],
        p,
    )
}  # Assign colors to conditions
```

```python
sleep_conditions = [
    "early_bsl_nrem",
    "early_rec_nrem_match",
    "early_rec_nrem",
    "late_rec_nrem",
]
```

## Cortical spindle measures

```python
var = "cxsp_rate_rel2total"
_df = df.loc[
    df["condition"].isin(sleep_conditions), ["subject", "experiment", "condition", var]
]
_df[var] = _df[var] * 100
fig, ax = plt.subplots(figsize=(1.25, 4))
g = sns.boxplot(
    data=_df,
    x="condition",
    y=var,
    fliersize=0,
    saturation=1.0,
    linewidth=1.0,
    order=sleep_conditions,
    ax=ax,
    hue="condition",
    palette=p,
)
sns.stripplot(
    data=_df, x="condition", y=var, color="k", s=1.5, ax=ax, order=sleep_conditions
)
ax.set_xticks(range(len(sleep_conditions)))
ax.set_xticklabels(["E", "L", "E", "L", "E", "L"][:-2])
plt.ylabel(None)
plt.xlabel(None)

fig.savefig(
    nb.get_project_file("figures/cxsp_rate_nrem_only.png"), bbox_inches="tight", dpi=600
)
```
