# Paper correlation plots

```python
experiment_names = ["Novelty", "Locomotion", "Dual"]
m_exp = ["o", "^", "P"]
p_exp = sns.color_palette("Dark2_r", 8)
p_exp = dict(zip(experiment_names, p_exp))
```

```python
def label_point(x, y, val, ax, **text_kwargs):
    a = pd.concat({"x": x, "y": y, "val": val}, axis=1)
    for i, point in a.iterrows():
        ax.text(point["x"] + 0.02, point["y"], str(point["val"]), **text_kwargs)


def make_corr_plot(
    measures: pd.DataFrame,
    x_cond: str,
    x_var: str,
    y_cond: str,
    y_var: str,
    sig_exp_lines: list[str] = [],
    trend_exp_lines: list[str] = [],
    sig_single_line: bool = False,
    trend_single_line: bool = False,
    aspect: float = 1.618,
    ht: float = 4.0,
    label_points=False,
) -> tuple[plt.Figure, plt.Axes]:
    new_xvar = f"{x_cond} {x_var}"
    new_yvar = f"{y_cond} {y_var}"
    x_df = measures.loc[
        measures["condition"] == x_cond, ["subject", "experiment", x_var]
    ].rename(columns={x_var: new_xvar})
    y_df = measures.loc[
        measures["condition"] == y_cond, ["subject", "experiment", y_var]
    ].rename(columns={y_var: new_yvar})
    df = pd.merge(x_df, y_df, on=["subject", "experiment"])

    fig, ax = plt.subplots(figsize=(aspect * ht, ht))
    if sig_single_line:
        sns.regplot(
            data=df,
            x=new_xvar,
            y=new_yvar,
            ax=ax,
            ci=False,
            robust=False,
            truncate=True,
            scatter=False,
            color="k",
            line_kws={"linewidth": 3},
        )
    if trend_single_line:
        sns.regplot(
            data=df,
            x=new_xvar,
            y=new_yvar,
            ax=ax,
            ci=False,
            robust=False,
            truncate=True,
            scatter=False,
            color="k",
            line_kws={"linewidth": 3, "ls": "--"},
        )
    sns.scatterplot(
        data=df,
        x=new_xvar,
        y=new_yvar,
        hue="experiment",
        hue_order=p_exp.keys(),
        palette=p_exp,
        style="experiment",
        style_order=p_exp.keys(),
        markers=m_exp,
        s=60,
        legend=False,
        ax=ax,
    )
    if label_points:
        label_point(df[new_xvar], df[new_yvar], df["subject"], ax, fontsize=6)
    for exp, color in p_exp.items():
        line_kws = {"ls": "--"} if exp in trend_exp_lines else {}
        if sig_single_line or trend_single_line:
            line_kws.update({"alpha": 0.33})
        if (exp in sig_exp_lines) or (exp in trend_exp_lines):
            sns.regplot(
                data=df.loc[df["experiment"] == exp],
                x=new_xvar,
                y=new_yvar,
                ax=ax,
                ci=False,
                robust=False,
                truncate=True,
                scatter=False,
                color=color,
                line_kws=line_kws,
            )
    ax.spines[["top", "right"]].set_visible(False)
    return fig, ax


def strip_axes(ax: plt.Axes) -> plt.Axes:
    ax.set_xticks([])
    ax.set_yticks([])
    ax.set_xlabel(None)
    ax.set_ylabel(None)
```

# Pre-post plots

```python
def prepost_plot(
    data: pd.DataFrame,
    x: str,
    order: list[str],
    y: str,
    connect: str,
    col: str,
    col_order: list[str],
    palette: list,
    figsize: tuple[float, float] = None,
):
    data = data.copy()

    data = data.loc[data[x].isin(order) & data[col].isin(col_order)]
    fig, axes = plt.subplots(
        1, len(col_order), sharex=True, sharey=True, figsize=figsize
    )
    if isinstance(axes, plt.Axes):
        axes = [axes]
    for i_col, col_val in enumerate(col_order):
        ax = axes[i_col]
        ax.set_title(col_val)
        col_df = data.loc[data[col] == col_val]
        for connector_val in col_df[connect].unique():
            connector_df = col_df.loc[col_df[connect] == connector_val].set_index(x)
            y_vals = connector_df.loc[list(order), y].values
            ax.plot(range(len(order)), y_vals, color="black", lw=1, zorder=0)
            ax.scatter(range(len(order)), y_vals, color=palette)
        ax.set_xlim(-0.5, len(order) - 0.5)
        ax.set_xticks(range(len(order)))
        ax.set_xticklabels(order)
        # ax.set_xlabel(x)
    axes[0].set_ylabel(y)
    return fig, axes
```
