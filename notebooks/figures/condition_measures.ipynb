# Pre-post plots

```python
def prepost_plot(
    data: pd.DataFrame,
    x: str,
    order: list[str],
    y: str,
    connect: str,
    col: str,
    col_order: list[str],
    palette: list,
    figsize: tuple[float, float] = None,
):
    data = data.copy()

    data = data.loc[data[x].isin(order) & data[col].isin(col_order)]
    fig, axes = plt.subplots(
        1, len(col_order), sharex=True, sharey=True, figsize=figsize
    )
    if isinstance(axes, plt.Axes):
        axes = [axes]
    for i_col, col_val in enumerate(col_order):
        ax = axes[i_col]
        ax.set_title(col_val)
        col_df = data.loc[data[col] == col_val]
        for connector_val in col_df[connect].unique():
            connector_df = col_df.loc[col_df[connect] == connector_val].set_index(x)
            y_vals = connector_df.loc[list(order), y].values
            ax.plot(range(len(order)), y_vals, color="black", lw=1, zorder=0)
            ax.scatter(range(len(order)), y_vals, color=palette)
        ax.set_xlim(-0.5, len(order) - 0.5)
        ax.set_xticks(range(len(order)))
        ax.set_xticklabels(order)
        # ax.set_xlabel(x)
    axes[0].set_ylabel(y)
    return fig, axes
```

```python
ewk_colors = [p["Early.EXT.Wake"], p["Late.EXT.Wake"]]
```

```python
# Extended Data Fig. 4A, right
plot_for_pub = True

fig, axes = prepost_plot(
    df,
    x="condition",
    order=("Early.EXT.Wake", "Late.EXT.Wake"),
    y="EMG",
    connect="subject",
    col="experiment",
    col_order=("Novelty", "Locomotion", "Dual"),
    palette=ewk_colors,
)
for ax, name in zip(axes, ["Novelty", "Locomotion", "Dual"]):
    if plot_for_pub:
        ax.set_title(None)
        ax.set_xlabel(None)
        ax.set_xticklabels([])
        ax.set_ylabel(None)
        ax.set_yticklabels([])
    else:
        ax.set_title(name)

# fig.savefig(nb.get_project_file("figures/emg_ewk_change.svg"), bbox_inches="tight")
# fig.savefig(nb.get_project_file("figures/emg_ewk_change.png"), dpi=600, bbox_inches="tight")
```
